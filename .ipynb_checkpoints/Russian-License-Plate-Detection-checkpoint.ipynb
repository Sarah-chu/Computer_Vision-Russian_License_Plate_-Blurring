# Object Detection Assessment Project Exercise

## Russian License Plate Blurring

Russians are famous for having some of the most entertaining DashCam footage on the internet but a lot of license plates in the footages were not blurred and thus the privacy of the driver was not protected. This mini-project is to help out and create a license plat blurring tool to address such issue.

### Method 
Haar Cascades algorithm will be used together with the  Russian license plate detector .xml file in OpenCV to blur the Russian license plates detected in an image.

------



### Import the libraries 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read in the car_plate.jpg file

```python
car = cv2.imread("../DATA/car_plate.jpg")
```

### Create a function that displays the image in a larger scale and correct coloring for 

```python
def display(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(img, cmap = 'gray')
```

```python
display(car)
```

### Load the haarcascade_russian_plate_number.xml file

```python
car_cascade = cv2.CascadeClassifier("../DATA/haarcascade_russian_plate_number.xml")
```

###  Create a function that takes in an image and draws a rectangle around the license plate

```python
def detect_plate(img):
    
    car_img = img.copy()
    
    # Detect the plate from the image
    car_rect = car_cascade.detectMultiScale(car_img, scaleFactor = 1.2, minNeighbors = 5)
    
    # Draw a rectangle for every detection
    for (x,y,w,h) in car_rect:
        cv2.rectangle(car_img, (x,y),(x+w, y+h), (0,0,255), 3)
        
    return car_img
        
```

```python
result = detect_plate(car)
```

```python
display(result)
```

### Create a function to blur the license plate


```python
def detect_and_blur_plate(img):
    
    car_img = img.copy()
    roi = img.copy()
    
    # Detect the plate from the image
    car_rect = car_cascade.detectMultiScale(car_img, scaleFactor = 1.2, minNeighbors = 5)
    
    # Blur every detection of license plate
    for (x,y,w,h) in car_rect:
        
        # Grab the Region of Interest(ROI)
        roi = roi[y:y+h, x:x+w]
        
        # Blur the ROI
        blurred_roi = cv2.medianBlur(roi, 7)
        
        # Merge the ROI with the original image
        car_img[y:y+h , x:x+w] = blurred_roi
    
    return car_img
```

```python
result = detect_and_blur_plate(car)
```

```python
display(result)
```
